```python
from keras.applications.vgg16 import VGG16
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input
import numpy as np
from keras.models import Model
import os
os.environ['CUDA_VISIBLE_DEVICES']='0'
```

```python
base_model = VGG16(weights='imagenet', include_top=True)
model = Model(inputs=base_model.input, outputs=base_model.get_layer('fc2').output)
```

```python
model.summary()
```

```python
train_pictures_path='../Data/train/train_rumor/'
```

```python
import numpy as np
from tqdm import tqdm
import os

train_names=os.listdir(train_pictures_path)
pbar = tqdm(total=len(train_names))

train_fea=[]
i=0
for name in train_names:
    pbar.update(1)
    img = image.load_img(train_pictures_path+name, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    feature = model.predict(x)
    train_fea.append(feature)
```

```python
train_fea=np.array(train_fea)
train_fea=train_fea[:,0,:]
print(train_fea.shape)
np.save('train_rumor_fea.npy',train_fea)
```

## 验证集测试

```python
test_la=np.concatenate([np.ones([448,1]),np.zeros([519,1])])
val_la=np.concatenate([np.ones([1000,1]),np.zeros([1000,1])])
train_la=np.concatenate([np.ones([7955,1]),np.zeros([7715,1])])
train_feature=np.concatenate([np.load('train_rumor_fea.npy'),np.load('train_nonrumor_fea.npy')])
val_feature=np.concatenate([np.load('val_rumor_fea.npy'),np.load('val_nonrumor_fea.npy')])
test_feature=np.concatenate([np.load('test_rumor_fea.npy'),np.load('test_nonrumor_fea.npy')])
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
train_x=train_feature
train_y=train_la
test_x=val_feature
test_y=val_la
#train_x,test_x,train_y,test_y = train_test_split(train_feature, train_labels, test_size=0.3, random_state=1)         
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

```python
clf_LR = LogisticRegression()
clf_LR.fit(train_x, train_y)
LR_prediction = clf_LR.predict(test_x)

print(accuracy_score(test_y, LR_prediction))
print(classification_report(test_y, LR_prediction, labels=[0.0, 1.0], target_names=['nonrumor', 'rumor'],digits=4))
```
