```python
import SE
import joblib
import json
import numpy as np
import time

MAX_COMMENTS = 100
```

```python
with open('../dataset/train.json', 'r') as f:
    train = json.load(f)
with open('../dataset/test.json', 'r') as f:
    test = json.load(f)
    
len(train), len(test)
```

```python
train[0].keys()
```

```python
def gen_content_se(pieces):
    arr = np.zeros((len(pieces), 55))

    for i, piece in enumerate(pieces):
        if i % 10 == 0:
            print('[{}] {}/{} done'.format(time.strftime("%Y-%m-%d %H:%M:%S",
                                                         time.localtime()), i, len(pieces)))

        text = piece['content']
        cut_words = piece['content_words'].split(' ')
        emotions_dict = piece['content_emotions']

        arr[i][:3] = SE.sentiment_words_count(cut_words)
        arr[i][3:32] = SE.dalianligong_arr(cut_words)
        arr[i][32:33] = SE.boson_value(cut_words)
        arr[i][33:38] = SE.emoticon_arr(text, cut_words)
        arr[i][38:43] = SE.symbols_count(text)
        arr[i][43:46] = SE.pronoun_count(cut_words)
        arr[i][46:47] = SE.negation_words_count(cut_words)
        arr[i][47:55] = SE.baidu_arr(emotions_dict)

    return arr
```

```python
train_content_se = gen_content_se(train)
test_content_se = gen_content_se(test)

train_content_se.shape, test_content_se.shape
```

```python
np.save('./data/train_emotion_{}.npy'.format(train_content_se.shape), train_content_se)
np.save('./data/test_emotion_{}.npy'.format(test_content_se.shape), test_content_se)
```
